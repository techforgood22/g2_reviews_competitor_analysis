```python
%%writefile g2_reviews.py
from botasaurus.browser import browser, Driver
import pandas as pd
from bs4 import BeautifulSoup
import xlsxwriter
import asyncio
import nest_asyncio
import re
import pandas as pd
nest_asyncio.apply()


def get_html_data(driver: Driver):
    placeholder, competitor=get_names()
    driver.google_get(f"https://www.g2.com/products/{placeholder}/reviews?page=5", bypass_cloudflare=True)
    heading = driver.get_text('.product-head__title [itemprop="name"]')
    driver.save_screenshot()

    soup = BeautifulSoup(driver.page_html, 'html.parser')
    review_divs = soup.find_all('div', itemprop='reviewBody')
    print("------------review_div ssssssss", review_divs)

    return review_divs

def concatenate_records(df):



    # Get the column names from the DataFrame
    #column_names = df.columns

    # Convert DataFrame to a dictionary with lists of values for each column
    data_dict = df.to_dict(orient='list')

    # Concatenate all values in a single row for each column
    concatenated_data = {col: ' '.join(map(str, values)) for col, values in data_dict.items()}

    # Create a new DataFrame with the concatenated values
    concatenated_df = pd.DataFrame([concatenated_data])
    writer = pd.ExcelWriter(f"concatenated_reviews.xls",engine="xlsxwriter")
    concatenated_df.to_excel(writer, index=False)
    writer.close()

    print(concatenated_df)
    return concatenated_df.to_dict(orient='records')

def get_names():
    placeholder = "gong"
    competitor = " Gong"
    return placeholder, competitor

def remove_special_characters(text):
    # Define a regular expression pattern to match special characters
    pattern = r'[^a-zA-Z0-9\s]'
    # Replace special characters with an empty string
    cleaned_text = re.sub(pattern, '', text)
    return cleaned_text


@browser
def g2_reviews( driver: Driver,data):
    placeholder, competitor=get_names()
    driver.google_get(f"https://www.g2.com/products/{placeholder}/reviews?page=5", bypass_cloudflare=True)
    heading = driver.get_text('.product-head__title [itemprop="name"]')
    driver.save_screenshot()

    soup = BeautifulSoup(driver.page_html, 'html.parser')
    competitor = "Gong"
    review_divs = soup.find_all('div', itemprop='reviewBody')
    print("------------review_div ssssssss", review_divs)
    data = []

    for review_div in review_divs:
        review_data = {}

        like_section = review_div.find('div', string=f'What do you like best about {competitor}?').find_next('p', class_='formatted-text').text.strip()
        dislike_section = review_div.find('div', string=f"What do you dislike about {competitor}?").find_next('p', class_='formatted-text').text.strip()
        problem_section = review_div.find('div', string=f"What problems is {competitor} solving and how is that benefiting you?").find_next('p', class_='formatted-text').text.strip()

        print("Like:", like_section)
        print("Dislike:", dislike_section)
        print("Problem:", problem_section)

        review_data['company'] = "gong"
        review_data['like'] = remove_special_characters(like_section)
        review_data['dislike'] = remove_special_characters(dislike_section)
        review_data['solved_problem'] = remove_special_characters(problem_section)
        data.append(review_data)
        print("Review Data: ", review_data)
        print("----------------------")


    df = pd.DataFrame(data, columns=["competitor", "like", "dislike", "solved_problem"])
    print("like", df["like"])
    print("----------------------", df)
    print("dislike", df["dislike"])
    print("solved_problem", df["solved_problem"])
    writer = pd.ExcelWriter(f"mettl.xls",engine="xlsxwriter")
    df.to_excel(writer, index=False)
    writer.close()
    return concatenate_records(df)


#scrape_heading_task()
g2_reviews()
```
